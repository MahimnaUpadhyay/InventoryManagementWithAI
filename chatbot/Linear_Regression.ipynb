```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

<h1>Importing Dataset</h1>

```python
dataset = pd.read_csv("../../../datasets/grocery_dataset/groc_inventory.csv")
dataset
```

```python
dataset.drop(columns=['Product_ID'])
```

```python
dataset['Manufacturing_Date'] = pd.to_datetime(dataset['Manufacturing_Date'])
dataset['Expiration_Date'] = pd.to_datetime(dataset['Expiration_Date'])
dataset['Days_Between'] = (dataset['Expiration_Date'] - dataset['Manufacturing_Date']).dt.days
dataset = dataset.drop(["Expiration_Date", "Manufacturing_Date"], axis=1)
```

<h1>Encoding the Data</h1>

```python
from sklearn.preprocessing import LabelEncoder

product_name = dataset['Product_Name']
category = dataset['Category']
status = dataset['Status']

le = LabelEncoder()

product_label_encoder = le.fit_transform(product_name)
category_label_encoder = le.fit_transform(category)
status_label_encoder = le.fit_transform(status)

dataset['Product_Name'] = product_label_encoder
dataset['Category'] = category_label_encoder
dataset['Status'] = status_label_encoder
```

<h1>Splitting the data into X and y variable</h1>

```python
features = ['Product_Name', 'Category', 'Days_Between', 'Unit_Price', 'Stock_Quantity', 'Unit_Sold', 'Status']
target = 'Unit_Sold'

X = dataset[features]
y= dataset[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

<h1>Applying Linear Regression</h1>

```python
dataset.dtypes
```

```python
from sklearn.metrics import r2_score, accuracy_score

linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)

y_pred = linear_regression.predict(X_test)

R2_Score = r2_score(y_test, y_pred)
print("R2 SCORE: ", R2_Score)
```
